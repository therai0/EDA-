## Introduction

The rapid growth of e-commerce platforms has generated massive amounts of customer transaction data. Understanding customer behavior is essential for improving customer retention, increasing sales, optimizing marketing strategies, and building recommendation systems.

This report presents an Exploratory Data Analysis (EDA) of the E-commerce Customer Behavior and Purchase Dataset. The dataset simulates customer transactions, demographic information, payment methods, return behavior, and customer churn within an online shopping environment.

The primary objective of this analysis is to uncover meaningful patterns in customer purchasing behavior and identify factors influencing customer churn and spending.
