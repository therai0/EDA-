## Bivariate Analysis

Body Mass is the main column here so we are analysing body mass column with each other columns

## Body mass and Bill Length
+ Positive relation between body mass and bill length

```python
plt.title("Relation between body mass and bill length")
sns.scatterplot(x=df["body_mass_g"],y=df["bill_length_mm"])
```

```python
plt.title("Relation between body mass and bill length")
sns.lineplot(x=df["body_mass_g"], y=df["bill_length_mm"])
```

```python
df["bill_length_mm"].corr(df["body_mass_g"])
```

## Bill Depth and Body mass 
+ Negative relation

```python
plt.title("Relation between bill depth and body mass")
sns.scatterplot(x=df["bill_depth_mm"],y=df["body_mass_g"])
```

```python
sns.lineplot(x=df["bill_depth_mm"],y=df["body_mass_g"])
```

```python
df["bill_depth_mm"].corr(df["body_mass_g"])
```

## Flipper length and Body Mass 
+ Perfect Positive relation

```python
plt.title("Relation between Flipper Length and Body Mass")
sns.scatterplot(x=df["flipper_length_mm"],y=df["body_mass_g"])
```

```python
plt.title("Relation between Flipper Length and Body Mass")
sns.lineplot(x=df["flipper_length_mm"],y=df["body_mass_g"])
```

```python
df["body_mass_g"].corr(df["flipper_length_mm"])
```

## Body Mass and Island
