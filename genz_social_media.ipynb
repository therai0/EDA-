```python
# cretain group 

# group 1 
#  age group between 13-18
grp_one= df[df["age"] <= 18]
grp_one["addiction_level"].value_counts()

plt.title("Pie Chart visualization of Age group(13-18)")
plt.pie(x=grp_one["addiction_level"].value_counts(),autopct="%1.1f%%",labels=["Medium","Low","High"])

```

```python
# group 2
#  age group between 19-22
grp_two= df.loc[df["age"].between(19,22)]
grp_two["addiction_level"].value_counts()

plt.title("Pie Chart visualization of Age group(18-22)")
plt.pie(x=grp_two["addiction_level"].value_counts(),autopct="%1.1f%%",labels=["Medium","Low","High"])
```

```python
# group 3
#  age group between 19-22
grp_three= df.loc[df["age"].between(23,27)]
grp_three["addiction_level"].value_counts()

plt.title("Pie Chart visualization of Age group(23-27)")
plt.pie(x=grp_three["addiction_level"].value_counts(),autopct="%1.1f%%",labels=["Medium","Low","High"])
```

## Addiction level and Primary platform
+ Most of the people are using instagram so people are highly addicted with insta

```python
a = pd.crosstab(df["addiction_level"],df["primary_platform"],normalize="index")

sns.heatmap(a,annot=True)

plt.title("Addiction level vs Primary platform")
plt.xlabel("Primary platform")
plt.ylabel("Addiction level")
```
